```python
# Q3.

# The ArithmeticError Exception is the base class for all errors associated with arithmetic operation.
# It includes :
#     ZeroDivisionError
#     OverflowError
#     FloatingPointError


# ZeroDivisionError
try:
    1/0
except ArithmeticError as e:
    print(e)
```

```python
# OverflowError

j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
except ArithmeticError as e:
    print(e)
```

```python
# Q4.

# LookupError is a base class in Python's exception hierarchy that represents errors that occur when you try to 
# access or look up a value in a sequence or mapping (like a list, dictionary, or tuple) and the key or index you 
# provide is not found or is invalid. It is not meant to be used directly; instead, it serves as a parent class for 
# more specific lookup-related exceptions like KeyError and IndexError.

# KeyError
try:
    my_dict = {'apple': 1, 'banana': 2, 'cherry': 3}
    print(my_dict['apple'])
    print(my_dict['grape'])
except KeyError as e:
    print('Key Error : ', e)
```

```python
# IndexError

try:
    my_list = [1, 2, 3, 4, 5]
    print(my_list[2])  
    print(my_list[10])  
except IndexError as e:
    print(e)
```

```python
# Q5.

# Import Error occurs when the imported module is not available. 
# ModuleNotFoundError is the derived exception class from ImportError.

try:
    import myname
    print(myname)
except ImportError as e:
    print(e)
```
